### Question 1  
Consider the following Python module:  
a = 0  
def b():  
global a  
a = c(a0  
def c(a):  
return a + 2   

After importing the module into the interpreter, you execute:  
- b()
- b()
- b()
- a
  
  ?
  
What value is displayed when the last expression (a) is evaluated? Explain your answer by indicating what happens in every executed statement.


### Solution

#### Initialization:

a = 0 initializes the global variable a to 0.

#### Function b:

The function b is defined to modify the global variable a. It uses the global keyword, indicating that it will modify the global variable a directly.
When b() is called, it sets a to the result of c(a), where c is another function.

#### Function c:

The function c takes a parameter a and returns a + 2 (increment the value of a by 2).

#### Execution Steps:

#### First call: b()

b() is executed, and inside it, c(a) is called with the current value of a, which is 0.
c(0) returns 0 + 2 = 2, so the value of a is updated to **2**.

#### Second call: b()

b() is executed again, and now c(a) is called with the updated value of a, which is 2.
c(2) returns 2 + 2 = 4, so the value of a is updated to **4**.

#### Third call: b()

b() is executed one more time, and now c(a) is called with the updated value of a, which is 4.
c(4) returns 4 + 2 = 6, so a is updated to **6**.

#### Final Value:
After executing b() three times, the final value of a is **6**

Thus, when the expression a is evaluated, the value displayed is: **6**

```python
def fileLength(filename):
    try:
        infile = open(filename)            # Open file in read mode
        content = infile.read()           # Read file content
        print(len(content))             # Calculate and print the length of the file
    except (FileNotFoundError, OSError):
        print(f"File {filename} not found.")  # Print error message if file not found or can not read as text file

# Calling the function to check the result
fileLength('midterm.py')
fileLength('idterm.py')
```

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the attributes of Marsupial and also:  

a. _extends_ the Marsupial __ init __ constructor to take, as input, the coordinates x and y of the Kangaroo object,  

b. _supports_ method jump that takes number values dx and dy as input and moves the kangaroo by dx units along the x-axis and by dy units along the y-axis, and   

c. _overloads_ the __ str __ operator so it behaves as shown below.  

- k = Kangaroo(0,0)
- print (k)
  
I am Kangaroo located at coordinates (0,0)
  
- k.put_in_pouch('doll')
- k.put_in_pouch('firetruck')
- k.put_in_pouch('kitten')

['doll', 'firetruck', 'kitten']  

- k.jump(1, 0)
- k.jump(1, 0)
- k.jump(1, 0)
- print(k)

I am a Kangaroo located at coordinates (3, 0)

```python
class Marsupial:
    def __init__(self):
        self.pouch = []  # Initialize an empty pouch

    def put_in_pouch(self, item):
        self.pouch.append(item)  # Add item to the pouch

    def pouch_contents(self):
        return self.pouch  # Return the contents of the pouch

class Kangaroo(Marsupial):
    def __init__(self, x, y):
        super().__init__()  # Call the constructor of Marsupial
        self.x = x  # Initialize x coordinate
        self.y = y  # Initialize y coordinate

    def jump(self, dx, dy):
        self.x += dx  # Update x coordinate
        self.y += dy  # Update y coordinate

    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"  

# Testing the given example:
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())  
k = Kangaroo(0, 0)
print(k)  

k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())  
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)
print(k)  
```

### Question 4  
Write function collatz() that takes a positive integer x as input and prints the Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying this rule to the previious number x in the sequence:   

        x = x/2  if x is even  

        x = 3x + 1  if x is odd  
    
Your function should stop when the sequence gets to number 1. Your implementation must be recursive, without any loops.  

- collatz(1)
  
  1
- collatz(10)
  
  10<br>
  5<br>
  16<br>
  8<br>
  4<br>
  2<br>
  1
  

```python
def collatz(x):
    print(x)  # Print the given number
    if x == 1:
        return  # Stop when reaching 1
    elif x % 2 == 0:  # Check if the given number is even
        collatz(x // 2)  # Call collatz with x / 2
    else:  # If given number is odd
        collatz(3 * x + 1)  # Call collatz with 3x + 1

try:
    userInput = int(input("Enter a positive integer: "))
    if userInput > 0:
        collatz(userInput)
    else:
        print("Please enter a positive integer.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

### Question 5  
Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.  

- binary(0)<br>
0
- binary(1)<br>
1
- binary(3)<br>
11
- binary(9)<br>
1001  

```python
def binary(n):
    if n < 0:
        raise ValueError("Input must be a non-negative integer.")
    
    if n == 0:
        print(0, end='')  
    elif n == 1:
        print(1, end='')  
    else:
        binary(n // 2)  # Recursive call with integer division by 2
        print(n % 2, end='')  # Print the remainder after division by 2

# Taking user input 
try:
    userInput = int(input("Enter a non-negative integer: "))
    if userInput >= 0:
        binary(userInput)
    else:
        print("Please enter a non-negative integer.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
    
```

### Question 6  
Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your impementation using w3c.html.  

- infile = open('w3c.html')
- content = infile.read()
- infile.close()
- hp = HeadingParser()
- hp.feed(content)
  
W3C Mission  
&nbsp;
Principles



```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.inHeading = False
        self.headingLevel = 0

    def handle_starttag(self, tag, attrs):
        # Check if the tag is a heading tag (h1 to h6)
        if tag in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            self.inHeading = True
            self.headingLevel = int(tag[1]) - 1  # Calculate indentation level
            
    def handle_endtag(self, tag):
        # Reset inHeading flag when a heading tag ends
        if tag in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            self.inHeading = False
            
    def handle_data(self, data):
        # Print the heading with the appropriate indentation
        if self.inHeading:
            print('  ' * self.headingLevel + data.strip())

# Example usage
try:
    infile = open('w3c.html')
    content = infile.read()    
    hp = HeadingParser()
    hp.feed(content)
except FileNotFoundError:
    print("The file 'w3c.html' was not found. Please check the file path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

### Question 7  
Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the path of a URL.  

- webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)
<pre>  
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html
      http://reed.cs.depaul.edu/lperkovic/csc242/test2.html
          http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
      http://reed.cs.depaul.edu/lperkovic/csc242/test3.html
          http://reed.cs.depaul.edu/lperkovic/csc242/test4.html
</pre>

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def webdir(url, depth, indent):
    # Base case: stop recursion if depth is 0
    if depth < 0:
        return

    # Print the URL with the appropriate level of indentation
    print(' ' * indent + url)

    # Try to retrieve and parse the page
    try:
        response = requests.get(url)
        # Check if the request was successful
        if response.status_code != 200:
            return
        
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all <a> tags that contain links
        for link in soup.find_all('a', href=True):
            # Get the href attribute (link)
            next_url = link.get('href')

            # Create an absolute URL by joining with the base URL
            next_url = urljoin(url, next_url)

            # Recursively call webdir on the next URL, with reduced depth and increased indentation
            webdir(next_url, depth - 1, indent + 4)
    
    except requests.RequestException as e:
        print(f"Error accessing {url}: {e}")
        
# Example Usage
webdir('http://www.angelfire.com/super/badwebs/', 2, 0)
```

### Question 8  
Write SQL queries on the below database table that return:  

    a) All the temperature data.  
    b) All the cities, but without repetition.  
    c) All the records for India.  
    d) All the Fall records.  
    e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters.  
    f) The city and country for which the average Fall temprature is above 20 degrees, in increasing temprature order.  
    g) The total annual rainfall for Cairo.  
    h) The total rainfall for each season.  

| City    | Country          | Season | Temperature (C) | Rainfall (mm) |
|---------|------------------|--------|-----------------|---------------|
| Mumbai  | India             | Winter | 24.8            | 5.9           |
| Mumbai  | India             | Spring | 28.4            | 16.2          |
| Mumbai  | India             | Summer | 27.9            | 1549.4        |
| Mumbai  | India             | Fall   | 27.6            | 346.0         |
| London  | United Kingdom    | Winter | 4.2             | 207.7         |
| London  | United Kingdom    | Spring | 8.3             | 169.6         |
| London  | United Kingdom    | Summer | 15.7            | 157.0         |
| London  | United Kingdom    | Fall   | 10.4            | 218.5         |
| Cairo   | Egypt             | Winter | 13.6            | 16.5          |

| Cairo   | Egypt             | Summer | 27.7            | 0.1           |
| Cairo   | Egypt             | Fall   | 22.2            | 4.5           |





```python
import sqlite3

# Create an in-memory SQLite database
connection = sqlite3.connect(':memory:')
cursor = connection.cursor()

# Create the Weather table
cursor.execute('''
CREATE TABLE Weather (
    City TEXT,
    Country TEXT,
    Season TEXT,
    Temperature REAL,
    Rainfall REAL
)
''')

# Insert the given values into the Weather table
data = [
    ('Mumbai', 'India', 'Winter', 24.8, 5.9),
    ('Mumbai', 'India', 'Spring', 28.4, 16.2),
    ('Mumbai', 'India', 'Summer', 27.9, 1549.4),
    ('Mumbai', 'India', 'Fall', 27.6, 346.0),
    ('London', 'United Kingdom', 'Winter', 4.2, 207.7),
    ('London', 'United Kingdom', 'Spring', 8.3, 169.6),
    ('London', 'United Kingdom', 'Summer', 15.7, 157.0),
    ('London', 'United Kingdom', 'Fall', 10.4, 218.5),
    ('Cairo', 'Egypt', 'Winter', 13.6, 16.5),
    ('Cairo', 'Egypt', 'Spring', 20.7, 6.5),
    ('Cairo', 'Egypt', 'Summer', 27.7, 0.1),
    ('Cairo', 'Egypt', 'Fall', 22.2, 4.5)
]

cursor.executemany('''
INSERT INTO Weather (City, Country, Season, Temperature, Rainfall)
VALUES (?, ?, ?, ?, ?)
''', data)

connection.commit()

# Function to execute a query 
def execute_query(query, singleValue=False):
    cursor.execute(query)
    rows = cursor.fetchall()
    for row in rows:
        if singleValue:
            print(f"{row[0]}")
        else:
            print(row)

# a) All the temperature data
print("a) All the temperature data:")
execute_query("SELECT Temperature FROM Weather", singleValue=True)
print("\n")

# b) All the cities, but without repetition
print("b) All the cities, but without repetition:")
execute_query("SELECT DISTINCT City FROM Weather", singleValue=True)
print("\n")

# c) All the records for India
print("c) All the records for India:")
execute_query("SELECT * FROM Weather WHERE Country = 'India'")
print("\n")

# d) All the Fall records
print("d) All the Fall records:")
execute_query("SELECT * FROM Weather WHERE Season = 'Fall'")
print("\n")

# e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters
print("e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters:")
execute_query('''
SELECT City, Country, Season 
FROM Weather 
GROUP BY City, Country, Season 
HAVING AVG(Rainfall) BETWEEN 200 AND 400
''')
print("\n")

# f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order
print("f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order:")
execute_query('''
SELECT City, Country 
FROM Weather 
WHERE Season = 'Fall' 
GROUP BY City, Country 
HAVING AVG(Temperature) > 20 
ORDER BY AVG(Temperature)
''')
print("\n")

# g) The total annual rainfall for Cairo
print("g) The total annual rainfall for Cairo:")
execute_query("SELECT SUM(Rainfall) AS Total_Annual_Rainfall FROM Weather WHERE City = 'Cairo'", singleValue=True)
print("\n")

# h) The total rainfall for each season
print("h) The total rainfall for each season:")
execute_query("SELECT Season, SUM(Rainfall) AS Total_Rainfall FROM Weather GROUP BY Season")
print("\n")

# Close the connection
connection.close()

```

### Question 9  
Suppose list words is defined as follows:  

words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']  

Write list comprehension expressions that use list words and generate the following lists:  

    a) ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE', 'LAZY', 'DOG']  
    b) ['the', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
    c) [3, 5, 5, 3, 5, 4, 3, 4, 3] (the list of lengths of words in list words).
    d) [['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN', 'brown', 5], ['FOX', 'fox', 3], ['JUMPS', 'jumps', 5], ['OVER', 'over', 4], 
       ['THE', 'the', 3], ['LAZY', 'lazy', 4], ['DOG', 'dog', 3]] (the list containing a list for every word of list words, where each 
       list contain the word in uppercase and    lowercase and the length of the word).  
    e) ['The', 'quick', 'brown', 'fox', 'jumps', over', 'the', 'lazy', 'dog'] (the list of words in list words containing 4 or more 
       characters)

```python
# Given list 
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# a) Uppercase all the words
upperCase = [word.upper() for word in words]
print("a)", upperCase)

# b) Lowercase all the words
lowerCase = [word.lower() for word in words]
print("b)", lowerCase)

# c) List of lengths of words
wordLength = [len(word) for word in words]
print("c)", wordLength)

# d) Lists with word in uppercase, lowercase, and its length
wordDetails = [[word.upper(), word.lower(), len(word)] for word in words]
print("d)", wordDetails)

# e) List of words with word length greater than or equal to 4
wordLen4 = [word for word in words if len(word) >= 4]
print("e)", wordLen4)
```
